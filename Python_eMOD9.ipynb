```python
dataJan = pd.read_excel('HDB Dataset - Jan.xlsx')
dataFeb = pd.read_csv('HDB Dataset - Feb.csv')
```

```python
dataJan
```

```python
dataJan = dataJan.drop_duplicates()
```

```python
dataJan
```

```python
dataFeb
```

```python
dataFeb = dataFeb.drop_duplicates()
```

```python
dataFeb
```

```python
dataFeb["resale_price"]. fillna(0, inplace = True)
dataFeb
```

```python
dataFeb.dtypes
```

```python
dataJan.dtypes
```

```python
dataJan["floor_area_sqm"] = dataJan["floor_area_sqm"].astype("float64")
dataJan["resale_price"] = dataJan["resale_price"].astype("float64")
```

```python
dataJan.dtypes
```

```python
dataFeb["month"] = dataFeb["month"].astype("datetime64")
```

```python
dataFeb.dtypes
```

```python
dataMerge = pd.merge(dataJan, dataFeb, how='outer')
```

```python
dataMerge
```

```python
dataMean = dataMerge.groupby("town", as_index=False).mean()
dataMean
```

```python
town = dataMean["town"]
floor = dataMean["floor_area_sqm"]
plt.title("Mean area per town")
plt.xlabel("Town")
plt.xticks(rotation='vertical')
plt.ylabel("Area")
plt.bar(town,floor,color = "#4CAF50")
plt.show()
```
